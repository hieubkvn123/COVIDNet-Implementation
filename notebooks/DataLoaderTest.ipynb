# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from dataloader import DataLoader

# Some constants
data_dir = 'data/covidx/train'
img_size = 480
val_ratio = 0.1
batch_size = 16
```

# Test data loader without augmentation

```python
# Create the loader
loader = DataLoader(data_dir, labels_as_subdir=True, one_hot=True,
            img_size=img_size, train_val_ratio=val_ratio, batch_size=batch_size,
                   random_noise=False)

# Load sample batch
batch = loader.get_train_batch()
images, labels = batch

# Visualize
fig, ax = plt.subplots(4,4, figsize=(16, 16))
for i in range(len(images)):
    row = i // 4
    col = i %  4
    
    image = images[i] * 127.5 + 127.5
    image = image.numpy().astype(np.uint8)
    
    ax[row][col].imshow(image)
    ax[row][col].set_title('Negative' if np.argmax(labels[i]) == 0 else 'Positive')
    
plt.show()
```
